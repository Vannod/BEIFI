#**Función objetivo**

#**Función 1**
$
Y=\frac{x^6}{60}-\frac{x^5}{50}+\frac{x^4}{2}+\frac{2x^3}{3}-3.2x^2-6.4x
$

```python
# Función objetivo
def F1(x):
  return - (x**6 / 60) - (x**5 / 50) + (x**4 / 2) + (2 * x**3 / 3) - (3.2 * x**2) - 6.4 * x

x = np.linspace(-10, 10, 1000)
plt.plot(x, F1(x), label="F(x)", color = 'royalblue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de la función F1(x)")
plt.grid(True)
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-50, 50)
plt.show()
```

##Método de la Bisección

```python
def d1F1(x):
  return -x**5/10 - x**4/10 + 2*x**3 + 2*x**2 - 6.4*x - 6.4

def Biseccion():
  a = 0
  b = 5
  epsilon = 1e-6

  cont = 0
  registro = []

  while(True):
    alpha = (a + b) / 2
    dF_a = d1F1(a)

    dF_alpha = d1F1(alpha)

    if(dF_a * dF_alpha < 0):
      b = alpha
    else:
      a = alpha

    Fa = F1(a)

    if(np.abs(a - b) < epsilon):
      print('.................................................')
      print("It: {:02d} a: {:.2f} F(a): {:.2f}".format(cont, a, Fa))
      break

    cont += 1
    registro.append([cont, a , Fa])
    print("It: {:02d} a: {:.2f} F(a): {:.2f}".format(cont, a, Fa))

  return registro

Resultados1 = Biseccion()
```

###Evaluación del método

```python
iteraciones = [r[0] for r in Resultados1]
valores_a = [r[1] for r in Resultados1]
valores_Fa = [r[2] for r in Resultados1]

plt.figure(figsize=(8, 6))
plt.plot(iteraciones, valores_a, marker='o', label='Valores para a')
plt.plot(iteraciones, valores_Fa, marker='x', label='Valor para F(a)')
plt.xlabel('Iteración')
plt.ylabel('Valor')
plt.title('Convergencia del método de Bisección')
plt.grid(True)
plt.legend()
plt.show()
```

##Método de Newton-Rapshon

```python
def d2F1(x):
  return -x**4/2 - 2*x**3/5 + 6*x**2 + 4*x - 6.4

def NewtonRaphson():
    a, b = 0, 5
    x = 1
    epsilon = 0.001

    cont = 0
    registro = []

    while True:
        # Calculo de f'(x) y f''(x)
        d1F_x = d1F1(x)
        d2F_x = d2F1(x)

        xprev = x
        x = xprev - d1F_x/d2F_x

        F_x = F1(x)

        cont = cont + 1
        registro.append([cont, x, F_x])

        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, x, F_x))

        if(np.abs(x - xprev) <= epsilon):
            print("----------------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, x, F_x))
            break

    return registro

Resultados1 = NewtonRaphson()
```

##Método de Secante

```python
def Secante():
    a, b = 0, 5
    epsilon = 1e-6

    cont = 0
    registro = []

    while True:
        # calcular f'(a)
        dF_a = d1F1(a)

        # calcular f'(b
        dF_b= d1F1(b)

        # calcular alpha
        alpha = b - dF_b / ((dF_b - dF_a)/(b - a))

        dF_alpha = d1F1(alpha)

        if dF_alpha > 0:
            b = alpha
        else:
            a = alpha

        F_alpha = F1(alpha)

        cont = cont + 1

        registro.append([cont, alpha, F_alpha])
        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha, F_alpha))

        if np.abs(dF_alpha) <= epsilon:
            print("------------------------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha, F_alpha))
            break

    return registro

Resultados1 = Secante()
```

##Método de Sección Dorada

```python
def GoldenSectionSearch():
    a = 0
    b = 5
    tau = 2 - 1.618033988
    epsilon = 1e-6

    cont = 0
    registro = []

    while(True):
        # Calcular alpha1 y alpha2
        alpha1 = a*(1 - tau) + b*tau
        alpha2 = a*tau + b*(1 - tau)

        # Calcular f(alpha1) y f(alpha2)
        F_alpha1 = F1(alpha1)
        F_alpha2 = F1(alpha2)

        if(F_alpha1 > F_alpha2):
            a = alpha1
        else:
            b = alpha2

        cont = cont + 1
        registro.append([cont, alpha1, F_alpha1])
        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha1, F_alpha1))

        if(np.abs(F_alpha1 - F_alpha2) < epsilon):
            print("-------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha1, F_alpha1))
            break

    return registro

Resultados1 = GoldenSectionSearch()
```

##Método de Fibonacci

```python
def Fibonacci(*params):
    a   = params[0]
    b   = params[1]
    k   = params[2]
    it  = params[3]
    reg = params[4]
    L   = params[5]
    n   = params[6]
    fs  = params[7]
    ep  = params[8]

    i_arriba = n - k + 1
    i_abajo = n + 1
    Lk = L*fs[i_arriba]/fs[i_abajo]

    xa = a + Lk
    xb = b - Lk

    F_xa = F1(xa)
    F_xb = F1(xb)

    if(F_xa > F_xb):
        a = xa
    else:
        b = xb

    if(np.abs(F_xa-F_xb) > ep):
        k = k + 1
        it = it + 1

        reg.append([it, xa, F_xa])
        print("It: {:02d} x: {:.6f} F(x): {:.2f}".format(it, xa, F_xa))

        return Fibonacci(a, b, k, it, reg, L, n, fs, ep)
    else:
        print("-------------------------------------------------------")
        print("It: {:02d} x: {:.6f} F(x): {:.2f}".format(it, xa, F_xa))

        return reg

def RunFibonacci():
    n = 100

    fs = [1, 1]
    for i in range(n):
        aux = fs[-1] + fs[-2]
        fs.append(aux)

    a, b, k = 0, 5, 2
    L = b - a

    return Fibonacci(a, b, k, 0, [], b-a, n, fs, 1e-6)

Resultados1 = RunFibonacci()
```

###Evaluación del método

```python
iteraciones = [r[0] for r in Resultados1]
valores_a = [r[1] for r in Resultados1]
valores_Fa = [r[2] for r in Resultados1]

plt.figure(figsize=(8, 6))
plt.plot(iteraciones, valores_a, marker='o', label='Valores para a')
plt.plot(iteraciones, valores_Fa, marker='x', label='Valor para F(a)')
plt.xlabel('Iteración')
plt.ylabel('Valor')
plt.title('Convergencia del método de Bisección')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def d1F2(x):
  return -2*x*(-x**6 - x**5 + 10*x**2 + 10*x + 12)/(x**2 + 6)**2 + (-6*x**5 - 5*x**4 + 20*x + 10)/(x**2 + 6)

def Biseccion():
  a = -1
  b = 1
  epsilon = 1e-6

  cont = 0
  registro = []

  while(True):
    alpha = (a + b) / 2
    dF_a = d1F2(a)

    dF_alpha = d1F2(alpha)

    if(dF_a * dF_alpha < 0):
      b = alpha
    else:
      a = alpha

    Fa = F2(a)

    if(np.abs(a - b) < epsilon):
      print('.................................................')
      print("It: {:02d} a: {:.2f} Fa: {:.2f}".format(cont, a, Fa))
      break

    cont += 1
    registro.append([cont, a , Fa])
    print("It: {:02d} a: {:.2f} F(a): {:.2f}".format(cont, a, Fa))

  return registro

Resultados2 = Biseccion()
```

```python
def F3(x):
  return np.sin(2*np.pi*x) / x - np.cos(3*np.pi*x) / x

x = np.linspace(-5, 5, 1000)
plt.plot(x, F3(x), label="F(x)", color = 'royalblue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de la función F3(x)")
plt.grid(True)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-10, 10)
plt.show()
```

##Método de la Bisección

```python
def d1F3(x):
  return 3*np.pi*np.sin(3*np.pi*x)/x + 2*np.pi*np.cos(2*np.pi*x)/x - np.sin(2*np.pi*x)/x**2 + np.cos(3*np.pi*x)/x**2

def Biseccion():
  a = -1
  b = 1
  epsilon = 1e-6

  cont = 0
  registro = []

  while(True):
    alpha = (a + b) / 2
    dF_a = d1F3(a)

    dF_alpha = d1F3(alpha)

    if(dF_a * dF_alpha < 0):
      b = alpha
    else:
      a = alpha

    Fa = F3(a)

    if(np.abs(a - b) < epsilon):
      print('.................................................')
      print("It: {:02d} a: {:.2f} Fa: {:.2f}".format(cont, a, Fa))
      break

    cont += 1
    registro.append([cont, a , Fa])
    print("It: {:02d} a: {:.2f} F(a): {:.2f}".format(cont, a, Fa))

  return registro

Resultados3 = Biseccion()
```

##Método de Newton-Rapshon

```python
def d1F3(x):
  return 3*np.pi*np.sin(3*np.pi*x)/x + 2*np.pi*np.cos(2*np.pi*x)/x - np.sin(2*np.pi*x)/x**2 + np.cos(3*np.pi*x)/x**2

def d2F3(x):
  return -4*np.pi**2*np.sin(2*np.pi*x)/x + 9*np.pi**2*np.cos(3*np.pi*x)/x - 6*np.pi*np.sin(3*np.pi*x)/x**2 - 4*np.pi*np.cos(2*np.pi*x)/x**2 + 2*np.sin(2*np.pi*x)/x**3 - 2*np.cos(3*np.pi*x)/x**3

def NewtonRaphson():
    a, b = -1, 1
    x = -0.01
    epsilon = 1e-6

    cont = 0
    registro = []

    while True:
        # Calculo de f'(x) y f''(x)
        d1F_x = d1F3(x)
        d2F_x = d2F3(x)

        xprev = x
        x = xprev - d1F_x/d2F_x

        F_x = F3(x)

        cont = cont + 1
        registro.append([cont, x, F_x])

        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, x, F_x))

        if(np.abs(x - xprev) <= epsilon):
            print("----------------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, x, F_x))
            break

    return registro

Resultados3 = NewtonRaphson()
```

##Método de Secante

```python
def Secante():
    a, b = -1, 1
    epsilon = 1e-6

    cont = 0
    registro = []

    while True:
        # calcular f'(a)
        dF_a = d1F3(a)

        # calcular f'(b
        dF_b= d1F3(b)

        # calcular alpha
        alpha = b - dF_b / ((dF_b - dF_a)/(b - a))

        dF_alpha = d1F3(alpha)

        if dF_alpha > 0:
            b = alpha
        else:
            a = alpha

        F_alpha = F3(alpha)

        cont = cont + 1

        registro.append([cont, alpha, F_alpha])
        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha, F_alpha))

        if np.abs(dF_alpha) <= epsilon:
            print("------------------------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha, F_alpha))
            break

    return registro

Resultados3 = Secante()
```

##Método de Sección Dorada

```python
def GoldenSectionSearch():
    a = -1
    b = 1
    tau = 2 - 1.618033988
    epsilon = 1e-6

    cont = 0
    registro = []

    while(True):
        # Calcular alpha1 y alpha2
        alpha1 = a*(1 - tau) + b*tau
        alpha2 = a*tau + b*(1 - tau)

        # Calcular f(alpha1) y f(alpha2)
        F_alpha1 = F3(alpha1)
        F_alpha2 = F3(alpha2)

        if(F_alpha1 > F_alpha2):
            a = alpha1
        else:
            b = alpha2

        cont = cont + 1
        registro.append([cont, alpha1, F_alpha1])
        print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha1, F_alpha1))

        if(np.abs(F_alpha1 - F_alpha2) < epsilon):
            print("-------------------------------------------------------")
            print("It: {:02d} x: {:.2f} F(x): {:.2f}".format(cont, alpha1, F_alpha1))
            break

    return registro

Resultados3 = GoldenSectionSearch()
```

##Método de Fibonacci

```python
def Fibonacci(*params):
    a   = params[0]
    b   = params[1]
    k   = params[2]
    it  = params[3]
    reg = params[4]
    L   = params[5]
    n   = params[6]
    fs  = params[7]
    ep  = params[8]

    i_arriba = n - k + 1
    i_abajo = n + 1
    Lk = L*fs[i_arriba]/fs[i_abajo]

    xa = a + Lk
    xb = b - Lk

    F_xa = F3(xa)
    F_xb = F3(xb)

    if(F_xa > F_xb):
        a = xa
    else:
        b = xb

    if(np.abs(F_xa-F_xb) > ep):
        k = k + 1
        it = it + 1

        reg.append([it, xa, F_xa])
        print("It: {:02d} x: {:.6f} F(x): {:.2f}".format(it, xa, F_xa))

        return Fibonacci(a, b, k, it, reg, L, n, fs, ep)
    else:
        print("-------------------------------------------------------")
        print("It: {:02d} x: {:.6f} F(x): {:.2f}".format(it, xa, F_xa))

        return reg

def RunFibonacci():
    n = 100

    fs = [1, 1]
    for i in range(n):
        aux = fs[-1] + fs[-2]
        fs.append(aux)

    a, b, k = -2, 2, 2
    L = b - a

    return Fibonacci(a, b, k, 0, [], b-a, n, fs, 1e-6)

Resultados3 = RunFibonacci()
```
